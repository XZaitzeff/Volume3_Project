```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing, SimpleExpSmoothing, Holt
import pandas as pd
```

We are trying the Holt-Winters method. It uses exponential smoothing to generate predictions as weighted averages of past observations. The weights decay exponentially, allowing us to weight recent observations more than older ones.

Parameters:
- $\alpha\in(0, 1)$: level parameter, values closer to one mean the model reacts quickly to recent changes and values close to zero mean the model changes slowly and puts more weight on long-run average
- $\beta$: trend parameter
- $\gamma$: seasonal parameter, can be additive or multiplicative depending on the data.

```python
# import data
product_filter = pd.read_csv("cleaned_produce.csv")["productname"] == "Broccoli Crowns"
data = pd.read_csv("cleaned_produce.csv")[product_filter]["averageretail"].to_numpy()
data
```

```python
# figure out how exponential smoothing class works
model = ExponentialSmoothing(data)
fit_model = model.fit(optimized=True)
forecast = fit_model.forecast(12)
forecast
```
